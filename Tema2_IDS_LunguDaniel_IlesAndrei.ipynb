```python
# 4. Sa se scrie o functie sum_divisors_digits care preia un numar natural strict pozitiv si returneaza
# suma cifrelor divizorilor sai. Intr-o alta functie black_hole se apeleaza in mod repetat 
# sum_divisors_digits pe numerele rezultate, pana cand se atinge un numar maxim de iteratii (implicit
# 1000) sau se ajunge la numarul 15. Puteti crea alte functii auxiliare.
# Exemplul 1: se pleaca de la ; divizorii sunt ; suma cifrelor divizorilor este 
# , acesta fiind rezultatul apelului sum_divisors_digits(15) . Functia black_hole
# se opreste deci dupa un singur apel al functiei sum_divisors_digits .
# Exemplul 2: se pleaca de la ; divizorii sunt iar sum_divisors_digits(21) este 
# ; se reia in functia black_hole apelul functiei sum_divisors_digits pana la
# epuizarea numarului de apeluri sau potentiala stabilizare in 15.

def divisors(n: int):
    """
    Args:
        n: an integer
    Returns:
        list of all divisors of the argument
    
    """
    lst=[]
    for index in range(1, n+1):
        if n%index==0:
            lst.append(index)
    return lst

def sum_digits(number: int):
    """
    Args:
        number: an integer
    Returns:
        sum of all the digits existing in the given argument
    
    """
    s=0
    while number!=0:
        digit=number%10
        s=s+digit
        number=number//10
    return s
        

def sum_divisors_digits(n: int):
    """
    Args:
        n: an integer
    Returns:
        sum of all digits of the given argument's divisors
    
    """
    lst=divisors(n)
    s=0
    for element in lst:
        s=s+sum_digits(element)
    
    return s

def black_hole():
    """
    Args: none
    
    Returns:
        prints the sum of divisors of a given number
        
    This function stops the iterations of calculating the sum of all digits
    of an integer's divisors if 1000 steps were taken or if the sum returns 15.
    
    """
    it=0
    stop=False
    x=int(input("Number = "))
    while it<1000 and stop==False:
        sum_div=sum_divisors_digits(x)
        print(f"\nIteration {it+1}")
        print(f"Sum of divisors = {sum_div}")
        if sum_div == 15:
            stop=True
            print("Stoped.")
        it+=1
        x=sum_div
black_hole()
        
```

```python
# 3. Sa se scrie o functie care primeste caile catre doua fisiere de tip text. Despre fiecare fiecare se stie ca are
# cate un numar pe linie (numarul de linii din fisiere este necunoscut apriori). Functia va returna un tuplu cu:
# lista de numere care apar in ambele fisiere si diferenta maxima care exista intre numerele din al doilea
# fisier si numerele din primul fisier. Numerele pot fi cu semn, intregi sau in virgula mobila (functia trebuie sa
# mearga in orice situatie). Incercati sa folositi functii predefinite din Python (sau NumPy) si collection
# comprehension.
# Exemplu: a.txt contine numerele 1, 2, 3, 4 (cate unul pe linie). b.txt contine 2, 3, 4, 10 (cate un numar pe linie).
# Functia va returna tuplul ([2, 3, 4], 9) . Numarul 9 este realizat de 10-1.


def file_tuple(file_name1 : str , file_name2: str):
    """
    Args:
        file_name1: first file name : str
        file_name2: second file name : str
    
    Returns:
        a list having the first element a list of the mutual 
        elements from both files; the second element from 
        the list returned is the maximum difference obtained 
        from the second file's maximum value and the 
        first one's minimum value.
    
    """
    f = open(file_name1, "r")
    g = open(file_name2, "r")
    tup1 = []
    tup2 = []
    for line in f:
        if type(line) is int or float:
            tup1.append(float(line))
        else:
            return 0
    for line in g:
        if type(line) is int or float:
            tup2.append(float(line))
        else:
            return 0
    lst = []
    for element in tup1:
        if element in tup2:
            lst.append(element)
    maxim = max(tup2)
    minim = min(tup1)

    return (lst, maxim - minim)

print(file_tuple("text1.txt", "text2.txt"))
```

```python
#1. Scrieti o functie care determina daca un numar este egal cu suma divizorilor sai, mai putin numarul insusi.

#Exemplu: divizorii lui 6 sunt 1, 2, 3, 6; suma celor mai mici decat 6 este 1+2+3=6=numarul de plecare. Numarul
#8 nu este in aceasta situatie: . Folositi aceasta functie pentru a afisa toate numere cu
#proprietatea ceruta pana la un dat.




def verif(nr: int,s: int):
    """
        Se determina daca un numar este egal cu suma divizorilor sai
        
        Parametrii:
            nr : numarul dat
            aux : copia numarului
            s : suma divizorilor
        
        Returneaza: 1 daca suma divizorilor este egala cu nr dat, 0 in caz contrar
    """
    aux: int = nr
    for i in range (1,aux-1):
        if (aux%i == 0):
            s += i
            aux /= i
    if (s == nr):
        return 1
    else:
        return 0
    
print (verif(6,0))
```

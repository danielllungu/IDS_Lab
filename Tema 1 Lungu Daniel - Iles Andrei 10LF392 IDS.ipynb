```python
# 2. Se da o lista de cuvinte. Sa se determine cuvintele unice (fara diferenta intre litere mici si mari) si de
# cate ori apare fiecare.



lista=['doi', 'doi', 'trei', 'Trei', 'un']
wordfreq=[]
for index in range(0, len(lista)):
    lista[index]=lista[index].lower()

freq=[]
    
for word in lista:
    wordfreq.append([word, lista.count(word)])

for element in wordfreq:
    if element not in freq:
        freq.append(element)

print(freq)
```

```python
# 8. Se da o lista de dictionare de forma:
# [
# {"nume": "Popescu",
# "gen": feminin,
# "varsta": 30,
# "salariu": 2000,
# "animal preferat": "pisica",
# },
# {"nume": "Irimia",
# "varsta": 40,
# "gen": "masculin"
# "salariu": 3000,
# "animal preferat": "caine"
# },
# # etc
# ]
# Sa se determine:
# A. salariul mediu, minim si maxim
# B. cei care au varsta intr-un anumit interval dat [a, b]
# C. care este animalul cu cele mai multe preferinte. Daca sunt mai multe astfel e animale, atunci se vor da
# toate
# D. cate persoane de gen feminin si respectiv masculin sunt in lista
# Nota: Toate dictionarele au aceleasi chei: nume etc.


lista=[{"nume": "Popescu","gen": "feminin","varsta": 30,
"salariu": 2000, "animal preferat": "pisica"
},
{"nume": "Irimia", "varsta": 40,
"gen": "masculin", "salariu": 3000,
"animal preferat": "caine"
}, {"nume":"Ionescu", "gen": "masculin", "varsta": 23, 
    "salariu":1900, "animal preferat":"iepure" },
      {"nume": "Lungu","gen": "masculin","varsta": 21,
"salariu": 2000, "animal preferat": "caine"
}, {"nume": "Axinte","gen": "feminin","varsta": 20,
"salariu": 1800, "animal preferat": "pisica"
}]

suma=0
nr=0

salarii=[]
animale=[]
freq_gen={"masculin":0, "feminin":0}
masc=0
fem=0
for persoana in lista:
    suma=suma+persoana["salariu"]
    nr=nr+1
    salarii.append(persoana["salariu"])
    animale.append(persoana["animal preferat"])
    if persoana["gen"]=="masculin":
        masc=masc+1
    elif persoana["gen"]=="feminin":
        fem=fem+1
    
freq_gen["masculin"]=masc
freq_gen["feminin"]=fem

print("Salariul mediu = ", suma/nr)
print("Salariul minim = ", min(salarii))
print("Salariul maxim = ", max(salarii))

varsta_a=int(input("a="))
varsta_b=int(input("b="))

print(f"Persoane cu varsta cuprinsa intre {varsta_a} si {varsta_b} ani :")
for pers in lista:
    if pers["varsta"]>=varsta_a and pers["varsta"]<=varsta_b:
        print(pers["nume"])

freq=[]
for animal in animale:
    freq.append(animale.count(animal))

animale_preferate=[]
preferat=max(freq)
for index in range(0, len(freq)):
    if freq[index]==preferat:
        if animale[index] not in animale_preferate:
            animale_preferate.append(animale[index])

print("Animal/e cu cele mai multe preferinte:\n", animale_preferate)

print("Persoane de gen feminin = ", freq_gen["feminin"])
print("Persoane de gen masculin = ", freq_gen["masculin"])
```

```python
# 7. Se da un dictionar cu persoane, si pentru fiecare persoana o lista de intervale de timp de
# disponibilitate. O disponibilitate incepe la o anumita ora fixa (ex: 17) sau ora fixa+30 de minute si dureaza
# exact 30 de minute. Exemplu:
# doodle = {
# "Popescu": ["13", "13.30", "16", "17"], # Popescu este disponibil intre: 13 si
# 13.30, 13.30-14, 16-16.30, 17-17.30
# "Ionescu": ["13", "13.30", "16", "17", "17.30", "11", "15"],
# "Maximilian": ["14", "14.30", "15", "15.30", "16", "16.30", "19", "19.30"]
# }
# Sa se determine toate intervalele in care toate persoanele sunt disponibile. Daca nu exista niciun astfel de
# interval, atunci sa se determine care sunt intervalele de timp in care numarul maxim de persoane poate
# participa. Daca sunt mai multe intervale pentru acest numar maxim de persoane, se vor da toate solutiile.
# In fiecare caz se da lista de persoane participante si intervalele orare.

doodle = { "Popescu": ["13", "13.30", "15", "17"], 
          "Ionescu": ["13", "13.30", "16", "17", "17.30", "11", "15"],
          "Maximilian": ["14", "14.30", "15", "15.30", "16",  "16.30", "19", "19.30"]
         }

                    
    
intervale1=[]

for nume1 in doodle:
    for nume2 in doodle:
        if nume1 is not nume2:
            intersectie=list(set(doodle[nume1]) & set(doodle[nume2]))
            if ((nume1, nume2), intersectie) not in intervale1 and ((nume2, nume1), intersectie) not in intervale1:
                intervale1.append(((nume1, nume2), intersectie))



modificat=1
while modificat==1:
    modificat=0
    for index1 in range(0, len(intervale1)-1):
        for index2 in range(index1+1, len(intervale1)):
            intersectie=list(set(intervale1[index1][1]) & set(intervale1[index2][1]))
            if len(intersectie) > 0:
                reuniune=list(set(intervale1[index1][0]) | set(intervale1[index2][0]))
                if (reuniune, intersectie) not in intervale1:
                    intervale1.append((reuniune, intersectie))
                    modificat=1

max_len=0
participanti_interval_maxim=[]
for element in intervale1:
    if len(element[0])>max_len:
        max_len=len(element[0])

for index in range(0, len(intervale1)):
    if len(intervale1[index][0])==max_len:
        gasit=0
        for nume in intervale1[index][0]:
           for element in participanti_interval_maxim:
               if nume in element[0]:
                   gasit=1
        if gasit==0:
           participanti_interval_maxim.append(intervale1[index])
           
print(participanti_interval_maxim)
        
```
